```python
import pygame
import tkinter as tk
from tkinter.filedialog import askdirectory
import os

my_unique_player = tk.Tk()
my_unique_player.title("Unique Music Player")
my_unique_player.geometry("450x350")

selected_directory = askdirectory()
os.chdir(selected_directory)
song_files = os.listdir()

playlist = tk.Listbox(my_unique_player, font="Helvetica 12 bold", bg='lightblue', selectmode=tk.SINGLE)
for item in song_files:
    position = 0
    playlist.insert(position, item)
    position += 1

pygame.init()
pygame.mixer.init()

def play_selected_song():
    selected_song = playlist.get(tk.ACTIVE)
    pygame.mixer.music.load(selected_song)
    current_song_var.set(f"Now playing: {selected_song}")
    pygame.mixer.music.play()

def stop_playing():
    pygame.mixer.music.stop()

def pause_playing():
    pygame.mixer.music.pause()

def resume_playing():
    pygame.mixer.music.unpause()

play_button = tk.Button(my_unique_player, width=5, height=3, font="Helvetica 12 bold", text="PLAY",
                        command=play_selected_song, bg="green", fg="white")
stop_button = tk.Button(my_unique_player, width=5, height=3, font="Helvetica 12 bold", text="STOP",
                        command=stop_playing, bg="red", fg="white")
pause_button = tk.Button(my_unique_player, width=5, height=3, font="Helvetica 12 bold", text="PAUSE",
                         command=pause_playing, bg="blue", fg="white")
resume_button = tk.Button(my_unique_player, width=5, height=3, font="Helvetica 12 bold", text="RESUME",
                          command=resume_playing, bg="orange", fg="white")

current_song_var = tk.StringVar()
current_song_label = tk.Label(my_unique_player, font="Helvetica 12 bold", textvariable=current_song_var)
current_song_label.pack()

play_button.pack(fill="x")
stop_button.pack(fill="x")
pause_button.pack(fill="x")
resume_button.pack(fill="x")
playlist.pack(fill="both", expand="yes")

my_unique_player.mainloop()
```
